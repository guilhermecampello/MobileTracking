```python
import pandas as pd

zones = pd.read_csv("zones.csv")
zones.describe()
```

```python
for zone in zones.groupby(['Position']):
    print
    print("{0},{1},{2},{3},{4},{5}".format(zone[1].MagY.mean(),zone[1].MagY.std,zone[1].MagY.min,zone[1].MagY.max))
```
